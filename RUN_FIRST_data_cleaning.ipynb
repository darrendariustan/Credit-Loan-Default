## 1.4 Correlation Calculations for Round 1 of Feature Selection

We then plot a correlation matrix to identify numerical columns that are highly correlated and then select one of them to be included in the model (e.g. funded_amnt and funded_amnt_inv are the same)

```python
# Determine correlation matrix for numerical features: We take absolute values to see the strength of the correlation.
corr_matrix = df[num_cols].corr().abs()
```

Next, we set a threshold to display highly correlated variables of > 85%. If too highly correlated, it would mean
1. They might be representing the same meaning in domain knowledge
2. Even if variables represent different meaning, the issue of multi-collinearity exist. This then creates redundancy in the model.

With redundancy, this means that:
1. Models learns noise instead of meaningful patterns due to excessive dependence on redundant features (Overfitting)
2. Models may distribute feature importance unevenly across features.

The Goal of the correlation calculations would then be to remove duplicated variables with correlations in one dimension (based on domain understanding representation) that is above our stated threshold below.

- To remove these duplicate variables, I will then categorize them manually through an experimental process. This means that I will let the dataset decide amongst a correlation pair which variable to drop, and eventually, I will know that the number of categories.
- In the end, I know that there are 12 variables which remain, in which any variable could arbitrarily represent these 12 different meanings.
- From here, to decipher what these 12 variables would represent, I then look through each variable and its pair (some of which are repeated), and then I put them in the same category.
- After putting these highly correlated variables in the same category, the category would have its own aggregate semantic meaning based on how similar the names of the variable are in the category. I then give this category a new name below.


Note:
- This approach I undertake below sounds like a precursor to Principal Component Analysis. However as of this assignment, we technically have not learnt PCA yet. And hence, I take a logical approach to determine the removal of highly correlated features.

```python
# Set the threshold for high correlation to 85%. This number is arbitrary and can be adjusted.
threshold = 0.85

# Find pairs of columns with correlation above the threshold
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            high_corr_pairs.append((col1, col2, corr_matrix.iloc[i, j]))

# Convert to DataFrame for easy viewing
high_corr_df = pd.DataFrame(high_corr_pairs, columns=["Column1", "Column2", "Correlation"])

high_corr_df.sort_values(by="Correlation", ascending=False)
```

#### Defining 'good' vs. 'bad' for both types of loans based on color coding from Business Case

```python
# good (0) = Fully Paid, bad (1) = Default or Charged Off
def closed_loan_labels(status):
    if status == "Fully Paid":
        return 0
    else:
        return 1

# good (0) = Current, bad (1) = Late (16-30 days), Late (31-120 days), In Grace Period
def open_loan_labels(status):
    if status == "Current":
        return 0
    else:
        return 1

df_closed["loan_default"] = df_closed["loan_status"].apply(closed_loan_labels)
df_open["loan_default"] = df_open["loan_status"].apply(open_loan_labels)
```
